```python
import torch
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.models import StaticEmbedding


# Without leaving sentence-transformers
# static_embedding = StaticEmbedding.from_distillation("Alibaba-NLP/gte-multilingual-base", device="cpu", pca_dims=256, trust_remote_code=True)
# static_embedding = StaticEmbedding.from_model2vec("minishlab/potion-base-32M")
static_embedding = StaticEmbedding.from_model2vec("minishlab/potion-multilingual-128M")
model = SentenceTransformer(modules=[static_embedding])
```

```python
import os
import torch

def generate_embeddings(skills_file_path, model, embeddings_cache_path):
    """Generates embeddings from a skills file and saves them to a cache file."""
    if not os.path.exists(skills_file_path):
        print(f"Error: The file '{skills_file_path}' was not found.")
        return None

    # Read skills from the file
    with open(skills_file_path, 'r') as f:
        skills_list = [line.strip() for line in f if line.strip()]

    # Encode the skills
    print(f"Encoding {len(skills_list)} skills...")
    skills_embeddings = model.encode(skills_list, convert_to_tensor=True)

    # Save the embeddings to a file
    print(f"Saving embeddings to cache: '{embeddings_cache_path}'")
    torch.save(skills_embeddings, embeddings_cache_path)
    return skills_embeddings

def load_embeddings(embeddings_cache_path):
    """Loads embeddings from a cache file."""
    if os.path.exists(embeddings_cache_path):
        print(f"Loading embeddings from cache: '{embeddings_cache_path}'")
        return torch.load(embeddings_cache_path)
    else:
        print(f"Error: Embeddings cache file not found at '{embeddings_cache_path}'")
        return None

```

```python
# Example usage (optional - you can remove this if you just want the functions)
skills_file_path = "skills.txt"
embeddings_cache_path = "skills_embeddings.pt"

def load_embedding_from_file(embeddings_cache_path=embeddings_cache_path, skills_file_path=skills_file_path):
    """Loads embeddings from a cache file."""
    # Try loading embeddings first
    skills_embeddings = load_embeddings(embeddings_cache_path)

    # If loading failed, generate embeddings
    if skills_embeddings is None:
        skills_embeddings = generate_embeddings(skills_file_path, model, embeddings_cache_path)

    if skills_embeddings is not None:
        print("Embeddings are ready to use.")

    with open(skills_file_path, 'r') as f:
        skills_list = [line.strip() for line in f if line.strip()]
    return skills_embeddings, skills_list
```

```python
all_skills_embeddings, all_skills_list = load_embedding_from_file()
```

```python
exact_matches = {}
remaining_skills_list = []

# Perform exact lowercase matching
for skill in skills_list:
    if skill.lower() in [s.lower() for s in all_skills_list]:
        # Find the exact match in all_skills_list (preserving original casing)
        for s in all_skills_list:
            if skill.lower() == s.lower():
                exact_matches[skill] = s
                break
    else:
        remaining_skills_list.append(skill)

# Now, perform similarity matching for the remaining skills
remaining_embeddings = model.encode(remaining_skills_list)
remaining_similarities = model.similarity(all_skills_embeddings, remaining_embeddings)

# Output the matches
print("Exact matches:")
for skill, matched_skill in exact_matches.items():
    print(f" - '{skill}': Exact match is '{matched_skill}'")

print("\nBest matches for remaining skills in skills_list (Top 3):")
for idx_j, sentence2 in enumerate(remaining_skills_list):
    print(f" - '{sentence2}':")
    top_k_scores, top_k_indices = torch.topk(remaining_similarities[:, idx_j], k=3)
    for i in range(3):
        skill1 = all_skills_list[top_k_indices[i]]
        score = top_k_scores[i].item()
        print(f"   - '{skill1}' with score {score:.4f}")
```
